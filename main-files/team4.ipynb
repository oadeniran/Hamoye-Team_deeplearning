```python
def rename_col(data, col_dict):
    data = data.rename(columns=col_dict)
    return data
```

```python
def replace_dash(data, col_dash):
    data[col_dash] = data[col_dash].replace('-', np.nan)

# replace_dash()
```

```python
def clean_data(data, col_rename):

    data = rename_col(data, col_rename)

    data.drop(index=0, axis=1, inplace=True)

    year = list(data.columns[:2])
    data[year] = data[year].replace('=', '', regex=True)
    data[year] = data[year].replace(' ', '', regex=True)

    # re.sub(r"[^a-zA-Z0-9-]","",text)
    # df1[col] = df1[col].apply(lambda x: clean(x) if type(x) == str else x)
    data[year[1]] = data[year[1]].apply(lambda x: re.sub(r"[^a-zA-Z0-9-]","",x) if type(x) == str else x)

    length_of_data = len(data)
    last_four = [x for x in range(length_of_data, length_of_data - 4, -1)]
    data.drop(index=last_four, axis=1, inplace=True)

    col_of_data_with_dash = list(data.columns[9:-1])
    data[col_of_data_with_dash] = data[col_of_data_with_dash].replace('-', np.nan)


    print('Information about missing data')
    print(data.isnull().sum() / len(data) * 100)

    special_character_col = list(data.columns[4:-1])

    for col in special_character_col:
        data[col] = data[col].apply(lambda x: clean(x) if type(x) == str else x)    

    data[col_of_data_with_dash] = data[col_of_data_with_dash].astype(float)
    data['Age'] = data['Age'].astype(float)


    return data
```
