# Aligning the column

```python
for d in [df_2018, df_2019, df_2020]:
    cols = []
    for column in d:
        if column[0].startswith('Unnamed'):
            cols.append(column[1])
        elif column[1].startswith('Unnamed'):
            cols.append(column[0])
        else:
            cols.append(column[0]+'_'+column[1])
    d.columns = cols
```

```python
# Replace the hyphens (-) with NaN

df_2018 = df_2018.replace("-", np.nan)
df_2019 = df_2019.replace("-", np.nan)
df_2020 = df_2020.replace("-", np.nan)
```

```python
# working on missing values- dropping them, especially for the machine learning model

df_2018.dropna(inplace = True)
df_2019.dropna(inplace = True)
df_2020.dropna(inplace = True)

for i in [df_2018, df_2019, df_2020]:
  print(i.shape)
```

```python
# rename some columns in df_2020- 'Rank in 2020' to '2020', 'rank in 2019' to '2019', to make it consistent with that of df_2018 and df_2019

cols = {
    "Rank in 2020": "2020",
    "Rank in 2019": "2019"
}
df_2020 = df_2020.rename(columns = cols)

df_2020
```
